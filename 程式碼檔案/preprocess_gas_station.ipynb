```python
import glob
import os
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import csv
import xml.etree.ElementTree as ET
```

```python
tree = ET.parse('../外部資料集/gas_station_2.xml')
root = tree.getroot()
```

```python
stations = []
for child in root:
    station_name = child.find('站名').text
    station_addr = child.find('地址').text
    station_lat = child.find('緯度').text
    station_lon = child.find('經度').text
    stations.append({
        'name': child.find('站名').text,
        'address': child.find('地址').text,
        'lat': child.find('緯度').text,
        'lon': child.find('經度').text,
    })
df_station1 = pd.DataFrame(stations)
df_station1
```

```python
df_station2 = pd.read_csv('../外部資料集/gas_station_1.csv')
df_station2.rename(columns={
    '加油站': 'name',
    '地址': 'address',
    'GPS定位座標北緯度': 'lat',
    'GPS定位座標東經度': 'lon',
}, inplace=True)
df_station2
```

```python
use_cols = ['name', 'address', 'lat', 'lon']
df_station = pd.concat([df_station1[use_cols], df_station2[use_cols]]).reset_index(drop=True)
```

```python
df_station['lat'] = df_station['lat'].astype(float)
df_station['lon'] = df_station['lon'].astype(float)
df_station.describe()
```

```python
df_station.to_csv('../官方資料集/external_data/gas_staion.csv', index=False, encoding='utf-8')
```
